```python
from langchain_core.messages import AIMessage

def get_role(fragment):
    if fragment.content.startswith("Resumen:"):
        return "Resumen anterior"
    elif isinstance(fragment, HumanMessage):
        return "Humano"
    elif isinstance(fragment, AIMessage):
        return "IA"
    else:
        return "Desconocido"


config = {"configurable": {"thread_id": "9999123457"}}

pregunta = "Como podria calcular mi nota final?"

# Verificar el último mensaje en la lista actual
last_message_type = get_role(output['recent_messages'][-1])

# Crear y agregar nuevo mensaje solo si no viola la alternancia
if last_message_type != "Humano":  # Si el último mensaje no es humano
    input_message = HumanMessage(content=pregunta)
    output['recent_messages'].append(input_message)
    output = chat_graph.invoke({"recent_messages": output['recent_messages'], "subject_id": subject_id}, config)
    for m in output['recent_messages'][-1:]:
        m.pretty_print()
else:
    print("No se puede agregar otro mensaje de usuario consecutivo.")
    print(output['recent_messages'][0])
    print(output['recent_messages'][1])
    print(output['recent_messages'][2])
    print(output['recent_messages'][3])
    print("Último mensaje:", output['recent_messages'][-1])
    
```
